The function below adds all the files found within the dataset into a list to be iterated through later.

```python
def createFileList(myDir, format='.png'):
    fileList = []
    print(myDir)
    #following loop walks through the files
    for root, dirs, files in os.walk(myDir, topdown=False):
        for name in files:
            if name.endswith(format):
                fullName = os.path.join(root, name)
                fileList.append(fullName)
    return fileList
```

The code below transcribes the png image files into ints from 0 to 255. Original dataset will expand all the way up to 6 GB for the csv. So I am utilizing only a fraction of the dataset. So in this case, I am splitting up the dataset into 4 pieces to be used. 


```python

def csv_creator(path, start, finish, trainList):
    counter= 0
    for file in trainList:

        label = file.split('/')
        #culling dataset because my computer keeps crashing when trying to load it into a dataframe
        skip_check = label[-1].split('_')
        skip_check = int(skip_check[-1][:4])
        print (file)
        if skip_check < start or skip_check > finish:
            continue
        print(counter)
        img_file = Image.open(file)
        label = file.split('/')
        label = label[-2].split('_')
        label = label[-1]
        print (label)
        # img_file.show()

        # get origanal image parameters...
        width, height = img_file.size
        format = img_file.format
        mode = img_file.mode

        # Make image Greyscale
        img_grey = img_file.convert('L')
        #img_grey.save('result.png')
        #img_grey.show()

        # Save Greyscale values
        value = np.asarray(img_grey.getdata(), dtype=np.int).reshape((img_grey.size[1], img_grey.size[0]))
        value = value[1::2]
        value = value.T[1::2]
        value = value.T
        value = value.flatten()
        value = gaussian_filter(value, sigma = 5)
        value = list(value)
        print (label)
        value.insert(0, label)

        #df.loc[counter]= value

        counter+= 1
        with open(path, 'a') as f:
            writer = csv.writer(f)
            writer.writerow(value)
```
